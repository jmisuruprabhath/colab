```python
import lxml
from bs4 import BeautifulSoup
from xlwt import *
from urllib.request import Request, urlopen
import pandas as pd
```

```python
req = Request(
    url = 'https://www.guessfactory.com/us/en/sale/women/view-all', 
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/192.168.56.1 Safari/537.36 QIHU 360SE'}
)

webpage = urlopen(req).read()
soup = BeautifulSoup(webpage, 'lxml')
products = soup.find_all('div', {'class': 'product-grid__column'})
```

```python
workbook = Workbook(encoding = 'utf-8')
table = workbook.add_sheet('data')
table.write(0, 0, 'Product Number')
table.write(0, 1, 'Product Name')
table.write(0, 2, 'New Price')
table.write(0, 3, 'Old Price')
table.write(0, 4, 'discount')
line = 1
```

```python
num = 0
df = pd.DataFrame(columns=['Product number', 'Product Name', 'New Price', 'Old Price', 'Discount'])
for product in products:
    product_cod = product.find('a')['href']
    product_code = product_cod.split("/")
    
    product_name = product.find('a', {'class': 'link product-tile__link'}).string.strip()
    new_price = product.find('span', {'class': 'price__red-color text-nowrap'}).text.strip().split('\n')[1]
    old_price = product.find('span', {'class': 'price__strike-through'}).string.strip()
    discount = product.find('span', {'class': 'price__red-color d-inline-block text-nowrap'}).text.strip().split(' ')[0].split('(')[1]
    # print(num, '\n', 'Product: ' + product_name.string.strip())
    # print('\n', product_code[len(product_code)-1])
    # print('\n', new_price)
    # print('\n', old_price.string.strip())
    # print('\n', discount)
    df = df.append(pd.DataFrame([[num,product_name,new_price,old_price,discount]], columns=['Product number', 'Product Name', 'New Price', 'Old Price', 'Discount']))
    num += 1
    table.write(line, 0, num)
    table.write(line, 1, product_name)
    table.write(line, 2, new_price)
    table.write(line, 3, old_price)
    table.write(line, 4, discount)
    line += 1
    workbook.save('GuessFactory_Products.xlsx')
```
